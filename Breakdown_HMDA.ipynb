```python
reduce_hmda_2023 = pd.read_csv('HMDA_2023_Reduce.csv')
reduce_hmda_2023.head()
```

```python
for idx, x in enumerate(reduce_hmda_2023.columns):
    if reduce_hmda_2023[x].isnull().any():
        print(f'dtype error in column "{x}" at index {idx}')
```

```python
reduce_hmda_2023['state_code'].unique().tolist()
```

replace nan to No Data

```python
reduce_hmda_2023['state_code'].replace(np.nan, 'No_Data', inplace = True)
reduce_hmda_2023['state_code'].unique()
```

```python
reduce_hmda_2023['county_code'].replace(np.nan, 0, inplace = True)
reduce_hmda_2023['county_code'].unique().tolist()
```

```python
reduce_hmda_2023['conforming_loan_limit'].unique()
reduce_hmda_2023['conforming_loan_limit'].replace(np.nan, 'Not Defined', inplace = True)
reduce_hmda_2023['conforming_loan_limit'].unique().tolist()
```

```python
reduce_hmda_2023['interest_rate'] = pd.to_numeric(reduce_hmda_2023['interest_rate'], errors = 'coerce')
reduce_hmda_2023['interest_rate'].replace(np.nan, reduce_hmda_2023['interest_rate'].mean(), inplace = True)
reduce_hmda_2023['interest_rate'].unique().tolist()
```
