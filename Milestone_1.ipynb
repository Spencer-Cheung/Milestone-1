```python
def get_fred_series_data(api_key, series):
    #url
    url = 'https://api.stlouisfed.org/fred/category/series?category_id=125&api_key=abcdefghijklmnopqrstuvwxyz123456&file_type=json'.format(series, api_key)
    #response
    response = requests.request("GET", url)
    return response
```

```python
def transform_series_response(response):
    latest_date = list(response.json()['meta']['data'].keys())[0]
    return pd.DataFrame(response.json()['meta']['data'].keys())[latest_date]
```

```python
def get_fred_data(param_list, start_date, end_date):
    df = pdr.DataReader(param_list, 'fred', start_date, end_date)
    return df.reset_index()
```
